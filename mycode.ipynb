```python
EPS=1e-6 #EPS stands for Epsilon. It is a tiny number (0.000001) .
#We add this EPS to our calculations later to avoid Division by Zero or log(0) errors, which would crash the code if a bin has zero Bads or zero Goods.

def compute_woe_iv_all(df,target='Default_Flag',bins=10,print_all=True):
    y=df[target].astype(int)
    results={}
    iv_rows={}
    
    for col in df.columns:
        if col==target:
            continue
        
        x=df[col]
        
        if pd.api.types.is_numeric_dtype(x) and x.nunique()>bins:
            try:
                binned_x=pd.qcut(x,q=bins,duplicates='drop')
            except Exception:
                binned_x=x.astype(str)
                
        else:
            binned_x=x.astype(str)
            
            
        d=pd.DataFrame({'Bin':binned_x,'Y':y})
        g=d.groupby('Bin')['Y']
        total=g.size()
        events=g.sum()
        non_events=total-events
        
        if events.sum()==0 or non_events.sum()==0:
            print(f"Skipping {col}: no events or non-events across its group.")
            continue
            
        woe_df=pd.DataFrame({
            'Total':total,
            'Events':events,
            'Non_Events':non_events
        })   
        
        woe_df['Dist_Event']=woe_df['Events']/woe_df['Events'].sum() # Dist_Event means Distribution of Events (i.e Defaults)
        woe_df['Dist_NonEvent']=woe_df['Non_Events']/woe_df['Non_Events'].sum()
        woe_df['WOE']=np.log((woe_df['Dist_NonEvent']+EPS)/(woe_df['Dist_Event']+EPS))
        woe_df['IV']=(woe_df['Dist_NonEvent']-woe_df['Dist_Event'])*woe_df['WOE']
        
        iv_total=woe_df['IV'].sum()
        
        results[col]=woe_df.reset_index().rename(columns={'Bin':col})
        iv_rows[col]=iv_total
        
        if print_all:
            print(f"/n===== {col} ===== ")
            print(results[col])
            print(f"Total IV for  {col} : {iv_total:.4f}")
            
            


#IV summary #
    iv_summary=pd.DataFrame(list(iv_rows.items()),columns=['Variable','IV']).sort_values('IV',ascending=False)
    def interpret(iv):
        if iv<0.02:return 'Not Predictive'
        if iv<0.10:return 'Weak'
        if iv<0.30:return 'Medium'
        if iv<0.50:return 'Strong'
        return 'Suspicious'
    
    iv_summary['Interpretation']=iv_summary['IV'].apply(interpret)
    
    print(iv_summary)
               
    return results,iv_summary
      

woe_tables, iv_summary = compute_woe_iv_all(df_clean, target='Default_Flag', bins=10, print_all=True)



            
```

```python
import pandas as pd
import numpy as np
## Age

df_clean['Age_bin']=np.select([df_clean['Age']<25,
                              (df_clean['Age']>=25) & (df_clean['Age']<=40),
                               df_clean['Age']>40 ],
                              ['High Risk (<25)', 'Medium Risk (25-40)', 'Low Risk (>40)'],default='Others'
                             )


##Income_Group
def bin_income(x):
    x=str(x).replace(" ","").lower()
    if x in ['<=15000','<=25000','<=35000']:
        return 'High Risk (<=35k)'
    elif x in ['<=50000','<=100000']:
        return 'Medium Risk (36k-100k)'
    elif x in ['>100000','>100k']:
        return 'Low Risk (>100k)'
    else:
        return 'Other'
    
df_clean['Income_bin']=df_clean['Income_group'].apply(bin_income)

## Occupation_Type

def bin_occ(x):
    x=str(x).strip().lower()
    if x in ['unemployed']:
        return 'High Risk (Unemployed)'
    elif x in ['private']:
        return 'Medium Risk (Private)'
    elif x in ['govt','self employed','self_employed','self-employed']:
        return 'Low Risk (govt/self employed)'
    else:
        return 'Other'
    
df_clean['Occ_bin']=df_clean['Occupation_type'].apply(bin_occ)

## Marital status

def bin_marital(x):
    x = str(x).strip().lower()
    if x == 'single':
        return 'High Risk (Single)'
    elif x in ['married', 'separated', 'seperated']:
        return 'Low Risk (Married/Separated)'
    else:
        return 'Other'

df_clean['Marital_bin'] = df_clean['Marital_Status'].apply(bin_marital)

## Housing Rent

df_clean['Housing_bin'] = df_clean['Housing_1_Own_2_Rent'].map({
    1: 'Low Risk (Own)',
    2: 'High Risk (Rent)'
})

##Employed_since_yrs

df_clean['EmpTenure_bin'] = np.select(
    [
        df_clean['Employed_since_yrs'] <= 3,
        (df_clean['Employed_since_yrs'] > 3) & (df_clean['Employed_since_yrs'] <= 10),
        df_clean['Employed_since_yrs'] > 10
    ],
    ['High Risk (<=3 yrs)', 'Medium Risk (4-10 yrs)', 'Low Risk (>10 yrs)'],
    default='Other'
)
    

## Occupation_Partner

def bin_occ_partner(x):
    x = str(x).strip().lower()
    if x in ['freelance', 'others']:
        return 'High Risk (Freelance/Others)'
    elif x == 'partime':
        return 'Medium Risk (Part-time)'
    elif x == 'fixed':
        return 'Low Risk (Fixed)'
    else:
        return 'Other'

df_clean['OccPartner_bin'] = df_clean['Occupation_Partner'].apply(bin_occ_partner) 


print(df_clean[['Age_bin','Income_bin','Occ_bin','Marital_bin',
                'Housing_bin','EmpTenure_bin','OccPartner_bin']].head(15)) 
```

```python
import pandas as pd
import numpy as np

target='Default_Flag'

coarse_vars=['Age_bin','Income_bin','Occ_bin','Marital_bin',
                'Housing_bin','EmpTenure_bin','OccPartner_bin']

def calc_woe_iv(df,feature,target):
    grouped=df.groupby(feature)[target].agg(['count','sum'])
    grouped.rename(columns={'count':'Total','sum':'Events'},inplace=True)
    grouped['Non_Events']=grouped['Total']-grouped['Events']

    grouped['Dist_Event']=grouped['Events']/grouped['Events'].sum()
    grouped['Dist_NonEvent']=grouped['Non_Events']/grouped['Non_Events'].sum()


    eps=1e-6 #EPS stands for Epsilon. It is a tiny number (0.000001) .
    #We add this EPS to our calculations later to avoid Division by Zero or log(0) errors, which would crash the code if a bin has zero Bads or zero Goods.

    grouped['WOE']=np.log( (grouped['Dist_NonEvent'] +eps)/(grouped['Dist_Event'] + eps))
    
    grouped['IV']=(grouped['Dist_NonEvent']-grouped['Dist_Event'])*grouped['WOE']
    grouped['IV']=grouped['IV'].round(6)
    
    total_iv=grouped['IV'].sum().round(4)
    
    return grouped.reset_index(),total_iv

iv_summary=[]

for  var in coarse_vars:
    table,total_iv=calc_woe_iv(df_clean,var,target)
    print(f"\n========")
    print(f"WOE/IV table for {var}")
    print(table)
    print(f"Total IV for {var}: {total_iv}")
    iv_summary.append({'Variable':var,'IV':total_iv})



iv_summary_df=pd.DataFrame(iv_summary)
print(iv_summary_df)

```

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor



def calc_woe_iv(df,feature,target):
    temp=df.groupby(feature)[target].agg(['count','sum'])
    temp['non_event']=temp['count']-temp['sum']
    temp['event_dist']=temp['sum']/temp['sum'].sum()
    temp['non_event_dist']=temp['non_event']/temp['non_event'].sum()
    eps=1e-6 #EPS stands for Epsilon. It is a tiny number (0.000001) .
    #We add this EPS to our calculations later to avoid Division by Zero or log(0) errors, which would crash the code if a bin has zero Bads or zero Goods.
    temp['WOE']=np.log((temp['non_event_dist']+eps)/(temp['event_dist']+eps))
    mapping=temp['WOE'].to_dict()
    return mapping

coarse_vars=['Age_bin','Income_bin','Occ_bin','Marital_bin',
                'Housing_bin','EmpTenure_bin','OccPartner_bin']

df_woe=df_clean.copy()
for var in coarse_vars:
    woe_map=calc_woe_iv(df_clean,var,'Default_Flag')
    df_woe[var+'_WOE']=df_clean[var].map(woe_map)
    
woe_cols=[v + '_WOE' for v in coarse_vars]
X=df_woe[woe_cols]

X_const = sm.add_constant(X) # this constant is 1 and it represents Intercept (alpha) in regression model ##
                             # This constant refers to value of dependent variable when all independent variables are zero
vif_data=pd.DataFrame()
vif_data['feature']=X_const.columns
vif_data['VIF']=[variance_inflation_factor(X_const.values,i) for i in range(X_const.shape[1])]

vif_data = vif_data[vif_data['feature'] != 'const'] ## VIF is designed for predictors and The intercept is not a predictor
                                                    ## so intercept's VIF has no modelling interpretation, we ignore it and drop it from the VIF table
print(vif_data)

## we will drop variable EmpTenure_bin_WOE  becoz its VIF >2
```

```python
import pandas as pd
import statsmodels.api as sm


woe_vars=['Age_bin_WOE','Income_bin_WOE','Occ_bin_WOE','Marital_bin_WOE',
                'Housing_bin_WOE','OccPartner_bin_WOE']

df_model=df_woe[woe_vars + ['Default_Flag']].copy()

print("Shape of modeling dataset:", df_model.shape)
print(df_model.head())

X=df_model[woe_vars]
y=df_model['Default_Flag']


X=sm.add_constant(X) # this constant is 1 and it represents Intercept (alpha) in regression model. #
                    # This constant refers to value of dependent variable when all independent variables are zero ##

logit_model=sm.Logit(y,X)
result=logit_model.fit()
## Statsmodels Logit reports McFadden’s Pseudo R² . So there are 3 types of Pseudo R² : McFadden , Cox & Snell and Nagelkerke.x

print(result.summary())
```

```python
import pandas as pd
import numpy as np

df_model['pred_prob']=result.predict(sm.add_constant(df_model[woe_vars]))

default=df_model[df_model['Default_Flag']==1]['pred_prob']
nondefault=df_model[df_model['Default_Flag']==0]['pred_prob']

concordant=0
discordant=0
ties=0

for i in default:
    for j in nondefault:
        if i>j:
            concordant+=1
        elif i<j:
            discordant+=1
        else:
            ties+=1
            
total_pairs=concordant+discordant+ties


concordance_perc=round((concordant/total_pairs)*100,2)
discordance_perc=round((discordant/total_pairs)*100,2)
ties_perc=round((ties/total_pairs)*100,2)

print("Concordant pair:",concordant)
print("Discordant pair:",discordant)
print("Ties pair:",ties)
print("Total pair:",total_pairs)

print("\n concordance (%):",concordance_perc)
print("\n discordance (%):",discordance_perc)
print("\n ties (%):",ties_perc)
```

```python
import pandas as pd
import numpy as np
from sklearn.metrics import roc_auc_score,roc_curve

def ks_statistics(y_true,y_pred_prob):
    data=pd.DataFrame({"y_true":y_true,"y_pred_prob":y_pred_prob})
    data=data.sort_values(by="y_pred_prob",ascending=False)
    
    data['cum_bad']=(data['y_true']==1).cumsum()/(data['y_true']==1).sum()
    data['cum_good']=(data['y_true']==0).cumsum()/(data['y_true']==0).sum()
    
    data['ks']=np.abs(data['cum_bad']-data['cum_good'])
    ks_value=data['ks'].max()
    ks_at=data.loc[data['ks'].idxmax(),"y_pred_prob"]
    
    return ks_value,ks_at,data


ks_value,ks_cutoff,ks_table=ks_statistics(y_true,y_pred_prob)
print(f"KS Stats:{ks_value:.4f} at cut_off{ks_cutoff:.4f}")
    
    
    
auc=roc_auc_score(y_true,y_pred_prob)
gini=2*auc-1
print(f"AUC:{auc:.4f}")
print(f"Gini coefficient:{gini:.4f}")

ks_table

pd_threshold = np.percentile(ks_table["y_pred_prob"], 80)
pd_threshold
```

```python
import pandas as pd
import numpy as np
from sklearn.metrics import roc_auc_score

num_bads= df_model['Default_Flag'].sum()
num_goods = len(df_model) - num_bads
base_odds = num_goods / num_bads

print(f"Number of Goods : {num_goods}")
print(f"Number of Bads  : {num_bads}")
print(f"Base Odds (G:B): {base_odds:.2f}")

PDO = 20

B = PDO / np.log(2)
print(f"B (slope parameter): {B:.4f}")

BASE_SCORE = 600

A = BASE_SCORE + B * np.log(base_odds)
print(f"A (offset parameter): {A:.4f}")

df_model['log_odds'] = np.log(
    df_model['pred_prob'] / (1 - df_model['pred_prob'])
)

df_model['credit_score'] = A - B * df_model['log_odds'] #higher log-odds -> higher PD -> higher risk 
                                                       # but banks want higher score=low PD=good borrower, so we subtract
                                                      # this "B * log_odds" to invert the direction .

df_model['credit_score'].describe()

df_model['score_decile'] = pd.qcut(df_model['credit_score'], 10)

score_validation = df_model.groupby('score_decile').agg(
    total_customers=('Default_Flag', 'count'),
    bad_rate=('Default_Flag', 'mean'),
    avg_score=('credit_score', 'mean')
).reset_index()

score_validation

auc_score = roc_auc_score(
    df_model['Default_Flag'],
    df_model['credit_score']
)
print(f"\n score_validation : {score_validation}")
print(f"\n AUC using PD   : {roc_auc_score(df_model['Default_Flag'], df_model['pred_prob']):.4f}")
print(f"\n AUC using Score: {auc_score:.4f}")
# AUC using PD   : 0.7678 but AUC using Score: 0.2322 bcoz we inverted the direction . 
# but this is normal becoz 1-0.7678 = 0.2322 
df_model

```

```python
import numpy as np
import pandas as pd
from scipy import stats
import statsmodels.api as sm

X=df_model[['Age_bin_WOE','Income_bin_WOE','Occ_bin_WOE','Marital_bin_WOE',
                'Housing_bin_WOE','OccPartner_bin_WOE']]

X=sm.add_constant(X)
df_model['pred_prob']=result.predict(X)

df_model['decile']=pd.qcut(df_model['pred_prob'],10,duplicates='drop')

h1_table=(
           df_model.groupby('decile',observed=False)
    .apply(lambda d: pd.Series({
        'total':len(d),
        'observed_1':d['Default_Flag'].sum(),
        'observed_0':(1-d['Default_Flag']).sum(),
        'expected_1':d['pred_prob'].sum(),
        'expected_0':(1-d['pred_prob']).sum()
        
    })
          )
    .reset_index()

)
h1_table['chi_1']=(h1_table['observed_1'] - h1_table['expected_1'])**2/h1_table['expected_1']
h1_table['chi_0']=(h1_table['observed_0'] - h1_table['expected_0'])**2/h1_table['expected_0']

HL_stat=h1_table['chi_1'].sum() + h1_table['chi_0'].sum() #Each decile has an error ;
                                            # so to answer “How bad is the model overall?” Total error = sum of all decile errors


df=h1_table.shape[0] - 2 # degrees of dreedom (g-2 where g=groups=10)
p_value=1-stats.chi2.cdf(HL_stat,df)

print(h1_table[['decile','observed_1','expected_1','observed_0','expected_0']])
print("\nHL Statistics:",round(HL_stat,2))
print("\n Degree of freedom:",df)
print("\np-value:",round(p_value,4))

```
